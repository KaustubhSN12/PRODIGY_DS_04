```python
import pandas as pd
from textblob import TextBlob

# Sentiment Analysis function
def get_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

# Apply sentiment analysis to each review
df['Sentiment'] = df['Public Opinion'].apply(get_sentiment)

# Filter negative reviews
negative_reviews = df[df['Sentiment'] == 'Negative']

# Count negative reviews for each brand
negative_reviews_count = negative_reviews['Brands'].value_counts()

# Print brands with more negative reviews
print("Brands with more negative reviews:")
print(negative_reviews_count.head(10))
```

```python
import pandas as pd

# Define lists of brands and game companies
brands_list = ['Facebook', 'Amazon', 'Google', 'Microsoft','Verizon', 'Nvidia', 'PlayStation5(PS5)', 'HomeDepot', 'johnson&johnson']
game_companies_list = ['MaddenNFL', 'NBA2K', 'FIFA', 'Fortnite', 'Overwatch', 'TomClancysGhostRecon',
                       'Hearthstone', 'PlayerUnknownsBattlegrounds(PUBG)', 'GrandTheftAuto(GTA)',
                       'TomClancysRainbowSix', 'Nvidia', 'Dota2', 'LeagueOfLegends', 'Battlefield',
                       'ApexLegends', 'RedDeadRedemption(RDR)', 'Borderlands', 'AssassinsCreed',
                       'CallOfDutyBlackopsColdWar', 'CS-GO', 'WorldOfCraft', 'Cyberpunk2077']


# Function to categorize brands
def categorize_brand(brand):
    if brand in brands_list:
        return 'Brand'
    elif brand in game_companies_list:
        return 'Game Company'
    else:
        return 'Other'

# Apply categorization to the Brands column
df['Brand Type'] = df['Brands'].apply(categorize_brand)

# Separate data into brands and game companies
brands_df = df[df['Brand Type'] == 'Brand']
game_companies_df = df[df['Brand Type'] == 'Game Company']

# Extract company names and their types
brands_info = brands_df[['Brands', 'Brand Type']].drop_duplicates()
game_companies_info = game_companies_df[['Brands', 'Brand Type']].drop_duplicates()

# Print company names and their types separately
print("Brands:")
print(brands_info.to_string(index=False))
print("\nGame Companies:")
print(game_companies_info.to_string(index=False))
```

**Respective Brand analysis**

```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt

# Sample data
data = {
    'Number': [1, 2, 3, 4, 5],
    'Brands': ['Amazon', 'Amazon', 'Other', 'Amazon', 'Other'],
    'Reviews': ['Great experience with Amazon!', 'Poor service from Amazon.', 'Other brand review.', 'Amazon products are excellent.', 'Not satisfied with Amazon delivery.'],
    'Public Opinion': ['Positive', 'Negative', 'Neutral', 'Positive', 'Negative']
}

# Convert sample data to DataFrame
df = pd.DataFrame(data)

# Filter the dataset to include only reviews related to Amazon
amazon_reviews = df[df['Brands'] == 'Amazon']

# Sentiment Analysis function
def get_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

# Apply sentiment analysis to each review
amazon_reviews['Sentiment'] = amazon_reviews['Reviews'].apply(get_sentiment)

# Calculate sentiment scores for Amazon
sentiment_counts = amazon_reviews['Sentiment'].value_counts()

# Visualize sentiment scores for Amazon
plt.bar(sentiment_counts.index, sentiment_counts.values)
plt.title('Sentiment Analysis for Amazon')
plt.xlabel('Sentiment')
plt.ylabel('Number of Reviews')
plt.show()
```
